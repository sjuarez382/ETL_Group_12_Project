```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
tv_shows_file = "Resources/tv_shows.csv"
tv_shows_df = pd.read_csv(tv_shows_file)
tv_shows_df.head()
```

```python
netflix_shows_df = tv_shows_df[['Title', 'Age', 'Netflix']]
netflix_shows_df.head()
```

```python
netflix_shows_df = netflix_shows_df[netflix_shows_df.Netflix != 0]
```

```python
netflix_shows_df.dropna(inplace=True)
```

```python
netflix_shows_df
```

```python
netflix_shows_df.loc[netflix_shows_df.Age == 'all', 'Age']= 'All'
```

```python
netflix_shows_df
```

```python
netflix_shows_df = netflix_shows_df.rename(columns={'Netflix':'Streaming_Service'})
```

```python
netflix_shows_df.loc[netflix_shows_df.Streaming_Service == 1, 'Streaming_Service']= 'Netflix'
```

```python
netflix_shows_df
```
